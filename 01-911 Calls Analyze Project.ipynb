# 911 Calls Analyze Project

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

** Checking the info() of the df **

```python
df.info()
```

** Checking the head of df **

```python
df.head()
```

** Top 5 zipcodes for 911 calls **

```python
df['zip'].value_counts().head(5)
```

** The top 5 townships for 911 calls **

```python
df['twp'].value_counts().head(5)
```

** Number of unique titles for call **

```python
df['title'].nunique()
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

** The most common Reason for a 911 call **

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason', data=df)
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

** Using seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column **

```python
sns.countplot(x='Day of Week', data=df, hue='Reason')
# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

**Same for Month:**

```python
sns.countplot(x='Month', data=df, hue='Reason')
# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

** Representation is missing some Months

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

** the count of calls per month. **

```python
byMonth['twp'].plot()
```

** create a linear fit on the number of calls per month.**

```python
sns.lmplot(x='Month', y='twp', data = byMonth.reset_index())
```

```python
df['Date'] = df['timeStamp'].apply(lambda t:t.date())
```

** A plot of counts of 911 calls on monthly basis.**

```python
df.groupby('Date').count()['twp'].plot()
```

** 3 separate plots with each plot representing a Reason for the 911 call**

```python
df[df['Reason'] == 'Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
```

```python
df[df['Reason'] == 'Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
```

```python
df[df['Reason'] == 'EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
```

____
** We need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week.**

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

** HeatMap **

```python
plt.figure(figsize=(10,5))
sns.heatmap(dayHour)
```

** clustermap **

```python
sns.clustermap(dayHour)
```

** Same plots but for months **

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(dayMonth)
```

```python
sns.clustermap(dayMonth)
```
